# Astrocyte/Radial Glia

```python
! pip install scanpy
! pip install scProject -U
import scProject
import scanpy as sc
patterns = sc.read_h5ad('patterns_anndata.h5ad')
dataset = sc.read_h5ad('test_target.h5ad')
dataset_filtered, patterns_filtered = scProject.projection_object.filterAnnDatas(dataset, patterns, 'gene_id')
```

Perfrom a regression with .001 weight of regularization and 1% lasso to encourage as many features to show and then we can weed them out by increasing the amount of lasso and regularization

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= [10, 12]
scProject.projection_object.NNLR_ElasticNet(dataset_filtered, patterns_filtered, 'ARG01', .001, .01)
scProject.projection_object.pearsonMatrix(dataset_filtered, patterns_filtered, 'CellType', 12, 'ARG01', 'PearsARG01', True)
```

```python
plt.rcParams['figure.figsize']= [12, 10]
scProject.projection_object.UMAP_Projection(dataset_filtered, 'CellType', 'ARG01', 'UMAPARG01', 20)
```

```python
scProject.projection_object.featurePlots(dataset_filtered, [3, 45, 64, 68, 78], 'ARG01', 'UMAPARG01')
```

These features seem like very strong markers except for feature 78 as it also has the lowest pearson value. If we up the lasso we will be able to get a better sense of the strong features.

```python
plt.rcParams['figure.figsize']= [10, 12]
scProject.projection_object.NNLR_ElasticNet(dataset_filtered, patterns_filtered, 'ARG99', .001, .99)
scProject.projection_object.pearsonMatrix(dataset_filtered, patterns_filtered, 'CellType', 12, 'ARG99', 'PearsARG99', True)
```

```python
plt.rcParams['figure.figsize']= [12, 10]
scProject.projection_object.UMAP_Projection(dataset_filtered, 'CellType', 'ARG99', 'UMAPARG99', 20)
```

```python
scProject.projection_object.featurePlots(dataset_filtered, [3, 45, 64, 68, 78], 'ARG99', 'UMAPARG99')
```

Features 3 and 78 clearly dropped out, which shows that while they may be expressed in astrocytes and radial glia they are not the core of the cell type as they are not as necessary in reconstructing the cell type as the other features that persisted. This is a perfect example of the model being forced to choose the most important features.

Lastly, lets up the regularization while keeping the lasso high to see what happens.

```python
plt.rcParams['figure.figsize']= [10, 12]
scProject.projection_object.NNLR_ElasticNet(dataset_filtered, patterns_filtered, 'ARGHigh', .0025, .99)
scProject.projection_object.pearsonMatrix(dataset_filtered, patterns_filtered, 'CellType', 12, 'ARGHigh', 'PearsARGHigh', True)
```

```python
plt.rcParams['figure.figsize']= [12, 10]
scProject.projection_object.UMAP_Projection(dataset_filtered, 'CellType', 'ARGHigh', 'UMAPARGHigh', 20)
```

```python
scProject.projection_object.featurePlots(dataset_filtered, [3, 45, 64, 68, 78], 'ARGHigh', 'UMAPARGHigh')
```

This clearly shows us that features 64 and 68 are the most robust markers of astrocytes/radial glia in the cortex. Lastly, let's check the highest expressed genes to understand the biology. Feature 45 plays a role so let's print those genes as well.

```python
print(scProject.projection_object.importantGenes(patterns_filtered, 45, .1), "Feature 45 Genes")
print(scProject.projection_object.importantGenes(patterns_filtered, 64, .5), "Feature 64 Genes")
print(scProject.projection_object.importantGenes(patterns_filtered, 68, .3), "Feature 68 Genes")
```
