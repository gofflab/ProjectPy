# Features for Microglia in P6 mouse Cortex

```python
! pip install scanpy
! pip install scProject -U
import scProject
import scanpy as sc
patterns = sc.read_h5ad('patterns_anndata.h5ad')
dataset = sc.read_h5ad('test_target.h5ad')
dataset_filtered, patterns_filtered = scProject.projection_object.filterAnnDatas(dataset, patterns, 'gene_id')
```

Low regularization low 1% lasso

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= [10, 12]
scProject.projection_object.NNLR_ElasticNet(dataset_filtered, patterns_filtered, 'MG01', .001, .01)
scProject.projection_object.pearsonMatrix(dataset_filtered, patterns_filtered, 'CellType', 12, 'MG01', 'MG01Pears', True)
```

```python
plt.rcParams['figure.figsize']= [12, 10]
scProject.projection_object.UMAP_Projection(dataset_filtered, 'CellType', 'MG01', 'UMAPMG01', 20)
```

```python
scProject.projection_object.featurePlots(dataset_filtered, [5, 6, 24, 57, 58, 75], 'MG01', 'UMAPMG01')
```

As expected from the Pearson plot, features 5 and 24 are the stronger markers of microglia in the P6 mouse cortex. Lets up the lasso to encourage sparsity and strengthen the features that are the real drivers.

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= [10, 12]
scProject.projection_object.NNLR_ElasticNet(dataset_filtered, patterns_filtered, 'MG99', .0075, .99)
scProject.projection_object.pearsonMatrix(dataset_filtered, patterns_filtered, 'CellType', 12, 'MG99', 'MG99Pears', True)
```

```python
plt.rcParams['figure.figsize']= [12, 10]
scProject.projection_object.UMAP_Projection(dataset_filtered, 'CellType', 'MG99', 'UMAPMG99', 20)
```

Feature Plots on original UMAP.

```python
scProject.projection_object.featurePlots(dataset_filtered, [5, 6, 24, 57, 58, 75], 'MG99', 'UMAPMG01')
```

Feature Plots on new UMAP

```python
scProject.projection_object.featurePlots(dataset_filtered, [5, 6, 24, 57, 58, 75], 'MG99', 'UMAPMG99')
```

While this is clearly over regularized some features 5, 6 and 75(because of how many nonzero cells) persist. Lets print out the highest expressed genes from the features of interest and see what's inside.

```python
print(scProject.projection_object.importantGenes(patterns_filtered, 5, .3), "Feature 5 Genes")
print(scProject.projection_object.importantGenes(patterns_filtered, 6, .3), "Feature 6 Genes")
print(scProject.projection_object.importantGenes(patterns_filtered, 24, .3), "Feature 24 Genes")
print(scProject.projection_object.importantGenes(patterns_filtered, 75, .3), "Feature 75 Genes")
```

In short, Feature 24 which is expressed in a fraction of microglia has high expression of C1qa, C1qb, C1qc genes. Feature 24 is not included in all of the microglia suggesting that there exists a subtype of microglia in the P6 mouse cortex that expresses C1qa through c in much higher amounts. Features 5, 6, and 75 do not have the C1qs expressed in their most important genes. This shows a subtyping of microglia in terms of expression of C1q expression.
